```python
import pandas as pd
import matplotlib.pyplot as plt

file_path = '/content/Registered Indian population by region and type of residence - Sheet1.csv'
df = pd.read_csv(file_path)

plt.figure(figsize=(12, 8))
plt.plot(df['Year'], df['Total a + b + c'], marker='o', linestyle='-', color='b')
plt.title('Number of Indian Immigrants Over the Years')
plt.xlabel('Year')
plt.ylabel('Total Bands (on reserve,on crown land and off reserve)')
plt.grid(True)
plt.savefig('/content/sample_data/figure1.png')
plt.show()
```

```python
import pandas as pd

df = pd.read_csv('/content/student info cannada - Sheet1 - student info cannada - Sheet1.csv.csv', thousands=',')

df_transposed = df.transpose()

df_transposed.columns = df_transposed.iloc[0]

df_transposed = df_transposed[1:]

df_transposed = df_transposed.dropna(axis=1)

fig, ax = plt.subplots(figsize=(12, 8))

for i, country in enumerate(df_transposed.columns):
    ax.bar(i, df_transposed[country].sum(), label=country, alpha=0.7)

ax.set_xlabel('Country')
ax.set_ylabel('Total')
ax.set_title('Study permit holders by country of citizenship')

ax.set_xticks(range(len(df_transposed.columns)))
ax.set_xticklabels(df_transposed.columns, rotation=45, ha='right')

ax.legend(loc='upper left', bbox_to_anchor=(1, 1))

plt.tight_layout()
plt.savefig('/content/sample_data/student.png')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

file_path = '/content/Registered Indian population in Canada  by region  - Sheet1.csv'

df = pd.read_csv(file_path)

df['Number'] = df['Number'].replace({',': ''}, regex=True).astype(int)

plt.figure(figsize=(10, 6))
plt.barh(df['Region'], df['Number'], color='skyblue')
plt.xlabel('Number')
plt.ylabel('Region')
plt.title('Population Distribution by Region')
plt.savefig('/content/sample_data/region.png')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

file_path = '/content/Indo-Canadian demography by religion  - Sheet1.csv'
df = pd.read_csv(file_path)

df['population %'] = pd.to_numeric(df['population %'].str.rstrip('%'))

plt.figure(figsize=(10, 6))
plt.bar(df['Religious group'], df['population %'], color='skyblue')
plt.xlabel('Religious group')
plt.ylabel('Population %')
plt.title('Religious Composition')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.savefig('/content/sample_data/religiouscomp.png')
plt.show()
```
